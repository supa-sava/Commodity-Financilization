# Making the actual plot

```python
# Calculate gross positions for each category
combined_df['Prod_Merc_Gross'] = (
    combined_df['Prod_Merc_Positions_Long_ALL'] + 
    combined_df['Prod_Merc_Positions_Short_ALL']
)

combined_df['Swap_Gross'] = (
    combined_df['Swap_Positions_Long_All'] + 
    combined_df['Swap__Positions_Short_All'] + 
    2 * combined_df['Swap__Positions_Spread_All']
)

combined_df['Money_Manager_Gross'] = (
    combined_df['M_Money_Positions_Long_ALL'] + 
    combined_df['M_Money_Positions_Short_ALL'] + 
    2 * combined_df['M_Money_Positions_Spread_ALL']
)

combined_df['Other_Rept_Gross'] = (
    combined_df['Other_Rept_Positions_Long_ALL'] + 
    combined_df['Other_Rept_Positions_Short_ALL'] + 
    2 * combined_df['Other_Rept_Positions_Spread_ALL']
)

combined_df['NonRept_Gross'] = (
    combined_df['NonRept_Positions_Long_All'] + 
    combined_df['NonRept_Positions_Short_All']
)
```

```python
# Calculate percentages
combined_df['Prod_Merc_Pct'] = combined_df['Prod_Merc_Gross'] / (2 * combined_df['Open_Interest_All'])
combined_df['Swap_Pct'] = combined_df['Swap_Gross'] / (2 * combined_df['Open_Interest_All'])
combined_df['Money_Manager_Pct'] = combined_df['Money_Manager_Gross'] / (2 * combined_df['Open_Interest_All'])
combined_df['Other_Rept_Pct'] = combined_df['Other_Rept_Gross'] / (2 * combined_df['Open_Interest_All'])
combined_df['NonRept_Pct'] = combined_df['NonRept_Gross'] / (2 * combined_df['Open_Interest_All'])
combined_df
```

```python
# Group by date and average across commodities
agg_df = combined_df.groupby('Report_Date_as_MM_DD_YYYY').agg({
    'Prod_Merc_Pct': 'mean',
    'Swap_Pct': 'mean',
    'Money_Manager_Pct': 'mean',
    'Other_Rept_Pct': 'mean',
    'NonRept_Pct': 'mean'
}).reset_index()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

plt.figure(figsize=(14, 8))

plt.plot(
    agg_df['Report_Date_as_MM_DD_YYYY'], 
    agg_df['Prod_Merc_Pct']*100, 
    label='Producer/Merchant/Processor/User', 
    color='blue'
)
plt.plot(
    agg_df['Report_Date_as_MM_DD_YYYY'], 
    agg_df['Swap_Pct']*100, 
    label='Swap Dealers', 
    color='green'
)
plt.plot(
    agg_df['Report_Date_as_MM_DD_YYYY'], 
    agg_df['Money_Manager_Pct']*100, 
    label='Money Managers', 
    color='red'
)
plt.plot(
    agg_df['Report_Date_as_MM_DD_YYYY'], 
    agg_df['Other_Rept_Pct']*100, 
    label='Other Reportables', 
    color='purple'
)
plt.plot(
    agg_df['Report_Date_as_MM_DD_YYYY'], 
    agg_df['NonRept_Pct']*100, 
    label='Non-Reportables', 
    color='gray', 
    linestyle='--'
)

plt.title('Trader Composition as % of Total Open Interest (2006-2025)', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('% of Open Interest', fontsize=12)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)

# Format x-axis for dates
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import rcParams

# agg_df = agg_df[agg_df['Report_Date_as_MM_DD_YYYY'] < '2015-01-01']


# Set better default styles
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Arial']
rcParams['axes.labelsize'] = 12
rcParams['axes.titlesize'] = 14

# Create figure
fig, ax = plt.subplots(figsize=(16, 9))

# Define colors and order (from bottom to top)
categories = [
    ('Producer/Merchant/Processor/User', '#e5ebf9'),  # Light blue (RGB 229, 235, 249)
    ('Swap Dealers', '#b9cb8c'),                     # Light green (RGB 185, 203, 140)
    ('Money Managers', '#7a9ed6'),                   # Dark blue (RGB 122, 158, 214)
    ('Other Reportables', '#e5ebf9'),                # Light blue (RGB 229, 235, 249)
    ('Non-Reportables', '#e3c76f')                   # Yellow (RGB 227, 199, 111)
]

# Create stacked area plot
ax.stackplot(agg_df['Report_Date_as_MM_DD_YYYY'],
            agg_df['Prod_Merc_Pct']*100,
            agg_df['Swap_Pct']*100,
            agg_df['Money_Manager_Pct']*100,
            agg_df['Other_Rept_Pct']*100,
            agg_df['NonRept_Pct']*100,
            colors=[col for _, col in categories],
            alpha=0.9,
            edgecolor='white',
            linewidth=0.3)

# Add reference line
ax.axhline(100, color='black', linestyle=':', alpha=0.5, linewidth=1)

# Format dates
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax.xaxis.set_major_locator(mdates.YearLocator())
plt.xticks(rotation=45, ha='right')

# Add labels and title
ax.set_title('Trader Composition as Percentage of Open Interest', 
           fontsize=16, pad=20)
ax.set_xlabel('Date', fontsize=12, labelpad=10)
ax.set_ylabel('Percentage of Open Interest', fontsize=12, labelpad=10)

# Create custom legend
handles = [plt.Rectangle((0,0),1,1, fc=col, alpha=0.9) for _, col in categories]
ax.legend(handles, [name for name, _ in categories],
         loc='upper left', bbox_to_anchor=(1, 1),
         frameon=False, fontsize=10)

# Add grid
ax.grid(True, axis='y', linestyle='--', alpha=0.5)
ax.set_axisbelow(True)
# Set y-axis ticks to increments of 25
ax.set_yticks(range(0, 101, 25))  # From 0 to 100 with steps of 25


# Remove spines
for spine in ['top', 'right']:
    ax.spines[spine].set_visible(False)

# Set limits
ax.set_ylim(0, 100)
ax.set_xlim(agg_df['Report_Date_as_MM_DD_YYYY'].min(),
          agg_df['Report_Date_as_MM_DD_YYYY'].max())

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(right=0.85)  # Make space for legend

# Add data source annotation
plt.annotate('Data Source: CFTC Commitment of Traders Reports',
            (0.5, -0.15),
            xycoords='axes fraction',
            ha='center',
            fontsize=9,
            color='#666666')

plt.show()
```
